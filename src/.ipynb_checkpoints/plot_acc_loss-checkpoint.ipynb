```python
"""
author: ouyang tianxiong
date: 2019/10/28
des: plot the trend curve of training loss and acc 
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(name):
    train_file = pd.read_csv(r"./chart/%s_train_loss_acc.csv"%name)
    validate_file = pd.read_csv(r"./chart/%s_validate_loss_acc.csv"%name)
    assert len(train_file) == len(validate_file), "size of train and validate mismatch" 
    plt.figure(figsize=[10,10], dpi=300)
    plt.subplot(211)
    plt.title("loss curve of model %s" %name)
    plt.plot(list(range(len(train_file))), list(train_file["train_loss"]),label="train_loss", color='r',linestyle="-")
    plt.plot(list(range(len(validate_file))), list(validate_file["validate_loss"]),label="validate_loss", color='b',linestyle="-.")
    plt.xlabel('epochs')
    plt.ylabel('loss value')
    plt.legend()
    plt.subplot(212)
    plt.title("accurate curve model %s" %name)
    plt.plot(list(range(len(train_file))), list(train_file["train_acc"]),label="train_acc",color='r',linestyle="-")
    plt.plot(list(range(len(validate_file))), list(validate_file["validate_acc"]),label="validate_acc",color='b',linestyle="-.")
    plt.xlabel('epochs')
    plt.ylabel('accurate rate')
    plt.legend()
    plt.show()
```

```python
plot("CNNCompact")
```

```python
plot("CNN1")
```

```python
plot("CNN2")
```
